7. Проверить гипотезу, что сотрудники с высоким окладом
проводят на работе больше времени, чем сотрудники с низким
окладом

Гипотеза H0 - сотрудники с высоким окладом
проводят на работе столько же времени, что и сотрудники с низким
окладом

```python
salary_low = df[df['salary'] == 'low']['average_montly_hours']
salary_high = df[df['salary'] == 'high']['average_montly_hours']
```

```python
import scipy.stats as stats
```

```python
result = stats.ttest_ind(salary_low, salary_high, equal_var=False)

a = 0.05

if (result.pvalue<a):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

Результат:сотрудники с высоким окладом проводят на работе столько же времени, что и сотрудники с низким окладом

8. Рассчитать следующие показатели среди уволившихся и не
уволившихся сотрудников (по отдельности):

● Доля сотрудников с повышением за последние 5 лет

● Средняя степень удовлетворенности

● Среднее количество проектов

```python
working= df[(df['left'] == 0)]
fired =  df[(df['left'] == 1)]
```

```python
table = []

table.append(
    {
    'Показатель': 'Доля сотрудников с повышением за последние 5 лет',
    'Уволившиеся': round(fired.loc[fired.promotion_last_5years==1,'promotion_last_5years'].count()/fired['promotion_last_5years'].count(),3),
    'Действующие': round(working.loc[working.promotion_last_5years==1,'promotion_last_5years'].count()/working['promotion_last_5years'].count(),3)
    }
)

table.append(
    {
    'Показатель': 'Средняя степень удовлетворенности',
    'Уволившиеся': round(fired.satisfaction_level.mean(), 3),
    'Действующие': round(working.satisfaction_level.mean(), 3)
    }
)

table.append(
    {
    'Показатель': 'Среднее количество проектов',
    'Уволившиеся': round(fired.number_project.mean(), 3),
    'Действующие': round(working.number_project.mean(), 3)
    }
)
table
```

9. Разделить данные на тестовую и обучающую выборки.
Построить модель LDA, предсказывающую уволился ли
сотрудник на основе имеющихся факторов (кроме department и
salary).
Оценить качество модели на тестовой выборки

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,[0,1,2,3,4,5,7]]
y = df['left']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
model = LinearDiscriminantAnalysis()
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
model.score(X_test, y_test) # Коэффициент детерминации - 0.83, а это означает, что модель получилась хорошая
```

```python
accuracy_score(y_test, model.predict(X_test)) # Точность модели высокая
```
